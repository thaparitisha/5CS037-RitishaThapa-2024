## Task - 3 - Outlier Detection:


1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 Ã— IQR rule.
3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.
4. Discuss the characteristics of these outliers and their potential impact on regional averages.

````python

# Define IQR for 'GDP per Capita' and 'Score'
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]


outliers_gdp = detect_outliers(southAsia_df, 'Log GDP per capita')
outliers_score = detect_outliers(southAsia_df, 'score')

# Combine outliers into a single DataFrame (avoiding duplicates)
southAsia_outliers = pd.concat([outliers_gdp, outliers_score]).drop_duplicates()

# Print the outlier countries
print("Outlier Countries based on GDP per Capita and Score:")
southAsia_outliers[['Country name', 'Log GDP per capita', 'score']]
````

````python
# Step 3: Scatter Plot Highlighting Outliers
plt.figure(figsize=(12, 7))

# Plot all countries
sns.scatterplot(data=southAsia_df, x='Log GDP per capita', y='score', color='gray', s=100, label='Non-Outliers')

# Highlight outliers
sns.scatterplot(data=southAsia_outliers, x='Log GDP per capita', y='score', color='red', s=100, label='Outliers')

plt.title('Outlier Detection: GDP per Capita vs Score')
plt.xlabel('GDP per Capita')
plt.ylabel('Score')
# plt.legend()
plt.show()
````
